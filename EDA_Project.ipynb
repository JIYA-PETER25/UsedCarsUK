# Sorting and  Comparison Operations

```python
# Calculating the average price in the order of their registration year

df.groupby("Registration_Year")["Price_in_dollar"].mean()
```

```python
# To get the average and maximum of price and mileage for each car model in every year

df.groupby(["Registration_Year","car_model"]).agg(
avg_price = ("Price_in_dollar", "mean"),
max_price = ("Price_in_dollar", "max"),
avg_mileage = ("Mileage(miles)", "mean"),
max_mileage = ("Mileage(miles)", "max"),

)
```

```python
# To get the no of groups

df.groupby(["Registration_Year","car_model"]).ngroups
```

```python
# To retrieve data of a particular case
# Retrieving the details of the Vauxhall Corsa with fuel type diesel

data_Corsa_diesel=df.groupby(["car_model","Fuel type"]).get_group(("Vauxhall Corsa","Diesel"))
data_Corsa_diesel.head()
```

```python
# Retrieving the details of cars that have fuel type as diesel and body type as pickup

df.loc[(df["Body type"]=="Pickup") & (df["Fuel type"]=="Diesel")]
```

```python
# Retrieving the details of cars that have automatic gearbox  and emission class Euro 2

df.loc[(df["Gearbox"]=="Automatic") & (df["Emission Class"]=="Euro 2")]
```

```python
# Displaying the details of car depending on doors

df.loc[(df["Doors"] < 3) | (df["Doors"] >= 5)]
```

# Graphical Representations

A boxplot to detect outliers in price

```python
sns.boxplot(x=df["Price_in_dollar"])
```

A boxplot to detect outliers in mileage

```python
sns.boxplot(x=df["Mileage(miles)"])
```

 Plotting a bar graph that gives the count of each car models

```python
df.car_model.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5),colormap="viridis")
plt.title("Number of each model cars")
plt.ylabel("No of cars")
plt.xlabel("Car Model")
```

Line plot depicting the trend of 'Price_in_dollar' over 'Registration_Year

```python
sns.lineplot(x="Registration_Year",y="Price_in_dollar",data=df)
plt.title("Average price in each year")
plt.xlabel("Year")
plt.ylabel("price")
plt.xticks(rotation=90)
plt.show()
```

Bar chart illustrating the distribution of different 'Fuel Types' in the dataset

```python
fuel_type_counts =df["Fuel type"].value_counts() 
plt.bar(fuel_type_counts.index, fuel_type_counts.values, color='blue')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Distribution of Fuel Types')
plt.xticks(rotation=90)
plt.show()
```

Exploring the distribution of the 'Seats' variable using a histogram

```python
sns.set_style("darkgrid")
sns.set_palette("crest")

df['Seats'].plot(kind='hist',bins=20)
plt.xlabel("Seats")
plt.ylabel("Count")
plt.title("Distribution of Seats")
plt.show()
```

Barplot displaying the top 60 car models by their respective prices, providing insights into the price distribution among different car models

```python
car_df=df[["car_model","Price_in_dollar"]].head(60)
df1=pd.DataFrame(car_df)
sns.barplot(y="car_model",x="Price_in_dollar",data=df1)
plt.title("price of each model")
plt.xlabel("price")
plt.ylabel("car model")
```

Correlation Heatmap of the Dataset 

```python
# Plotting heatmap to know how the variables are correlated

plt.figure(figsize=(10,5))
h=df.corr()
sns.heatmap(h,cmap="BrBG",annot=True)
```

Scatter plot depicting the relationship between Car Price in dollars and Mileage in miles

```python
x=df["Price_in_dollar"]
y=df["Mileage(miles)"]
plt.scatter(x,y,color="hotpink",s=5)
plt.show()
```

A piechart for the emission class

```python
p=df["Emission Class"].value_counts()
mylabels = ["Euro 5","Euro 6","Euro 4","Euro 3","Euro 2","Euro 1"]
myexplode=[0.1,0.04,0,0,0,0]
plt.pie(p,labels=mylabels,shadow=True,explode=myexplode,autopct='%.0f%%')
plt.legend(title="Emission Class",)
plt.show()
```

Boxplot showing the distribution of car prices ('Price_in_dollar') across different body types in the dataset

```python
sns.boxplot(x="Body type",y="Price_in_dollar",data=df)
plt.show
```

Four subplots presenting histograms for various vehicle attributes, including 'Fuel type,' 'Body type,' 'Engine,' and 'Gearbox' 

```python
plt.figure(figsize=(12, 8))

# First subplot
plt.subplot(221)
sns.histplot(x="Fuel type", data=df)
plt.title("Histogram of Fuel type")

# Second subplot
plt.subplot(222)
sns.histplot(x="Body type", data=df)
plt.title("Histogram of Body type")

# Third subplot
plt.subplot(223)
sns.histplot(x="Engine", data=df)
plt.title("Histogram of Engine")

# Fourth subplot
plt.subplot(224)
sns.histplot(x="Gearbox", data=df)
plt.title("Histogram of Gearbox")
plt.xticks(rotation=45)

# Adjust spacing between subplots
plt.tight_layout()

# Display the subplots
plt.show()
```
